```python
ain = """start-A
start-b
A-c
A-b
b-d
A-end
b-end"""
l = ain.split('\n')
l = [i.split('-') for i in l]
```

```python
def find(tofind):
    found = []
    for entry in range(len(l)):
        if l[entry][0] == tofind:
            found.append([entry, 0])
        if l[entry][1] == tofind:
            found.append([entry, 1])
    return found
s = find('start')

tv = s
ways = 0

allways = []
ends = find('end')
ends = [e[0] for e in ends]

def dfs(index, fr, sv, w, wl, usedDouble):
    global allways
    to = (fr + 1) % 2
    wl += [l[index][fr]]
    if l[index][to] == 'end':
        wl += [l[index][to]]
        # print('@')
        # print(wl)
        allways.append(wl)
        return w
    to = (fr + 1) % 2
    # print(to)
    alreadyVis = l[index][to].islower() and l[index][to] in sv
    if (alreadyVis and usedDouble) or l[index][to] == 'start':
        # print('done')
        return 0
    elif alreadyVis:
        # print('here')
        usedDouble = True

    goingto = l[index][to]
    sv.add(goingto)
    # print(goingto)
    tos = find(goingto)
    tot = 0
    # print(to)
    # for t in tos:
    #     if t[]
    for t in tos:
        # print(f'from {l[index][fr]} to {l[t[0]][t[1]]}')
        wlc = deepcopy(wl)
        svc = deepcopy(sv)
        tot+= dfs(t[0],t[1],svc, w, wlc, usedDouble)
    return tot

for st in s:
    ways += dfs(st[0],st[1], set(['start']), 1, [], False)

print(len(allways))
allways

```
