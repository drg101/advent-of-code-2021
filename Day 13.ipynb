```python
from copy import deepcopy
import numpy as np
ain = open("input.txt", "r").read()

l = ain.split('\n\n')
ins = l[0].split('\n')
ins = [[int(i.split(',')[0]),int(i.split(',')[1])] for i in ins]
# l = [i.split('-') for i in l]
#l = ain.split(',')
#l = [int(j) for j in l]

# m = []
# for line in l:
#     m.append(list(map(int, list(line))))

# p = print

folds = l[1].split('\n')
ins
```

```python
folds
```

```python
def foldX(ins2, foldat):
    for dot in ins2:
        if dot[0] > foldat:
            dot[0] = foldat - (dot[0] - foldat)
    ss = [f'{e[0]},{e[1]}' for e in ins2]
    ss = set(ss)
    ss = list(ss)
    ss = [[int(r.split(',')[0]),int(r.split(',')[1])] for r in ss]
    return ss

def foldY(ins2, foldat):
    for dot in ins2:
        if dot[1] > foldat:
            dot[1] = foldat - (dot[1] - foldat)
    ss = [f'{e[0]},{e[1]}' for e in ins2]
    ss = set(ss)
    ss = list(ss)
    ss = [[int(r.split(',')[0]),int(r.split(',')[1])] for r in ss]
    return ss

ins2 = deepcopy(ins)
for fold in folds:
    instruct = fold.split('=')
    ax = instruct[0][-1]
    line = int(instruct[1])
    print(ax,line)
    if ax == 'x':
        ins2 = foldX(ins2, line)
    else:
        ins2 = foldY(ins2, line)
#     print(ins2)
    print(len(ins2))

arr = np.array(ins2)
```
