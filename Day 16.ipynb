```python
from collections import defaultdict 
from copy import deepcopy
import numpy as np
import matplotlib.pyplot as plt
ain = open("input.txt", "r").read()
# ain = 'F600BC2D8F'
```

```python
scale = 16 # equal to hexadecimal

b = bin(int(ain, scale))[2:].zfill(len(ain)*4)
# By changing the parameter of the zfill function we allow for any length of
# hexadecimal code, which is more useful when using this code.
```

```python
vnumsum = 0

def multiplyList(myList) :
    result = 1
    for x in myList:
         result = result * x
    return result

def processPacket(b):
    global vnumsum
    packetVersion = int(b[:3], 2)
#     vnumsum = vnumsum + packetVersion
#     print(f'running sum {vnumsum}')
    packetID = int(b[3:6], 2)
    print(f'packet version {packetVersion}, id {packetID}')

    lengthTypeID = int(b[6:7], 2)
    numbits = 11
    if lengthTypeID == 0:
        numbits = 15

    literal = packetID == 4
    literalContent = ''
    if literal:
        bnumber = ''
        for i in range(0, 90000, 5):
            done = False
            chunk = b[6 + i:6 + i + 5]
            bnumber += chunk[1:5]
            print(chunk)
            if chunk[0] == '0':
                break
        print(f'literal {int(bnumber,2)}')
        return int(bnumber, 2), 6 + i + 5

    lengthOrNumOfSubPackets = int(b[7:7 + numbits], 2)
    print(packetVersion, packetID, lengthTypeID, lengthOrNumOfSubPackets, b)
    packetReturns = []
    done = False
    startbit = 7 + numbits
    packetCount = 0
    packetLength = 0
    while not done:
        print(done)
        ret, endCol = processPacket(b[startbit:])
#         packetReturns.append(ret)
        print('packet returned', ret, endCol)
        packetReturns += [ret]
        startbit += endCol
        packetCount += 1
        packetLength += endCol
        print(f'plength: {packetLength}, id: {lengthTypeID}')
        if lengthTypeID == 1 and packetCount >= lengthOrNumOfSubPackets:
            done = True
        if lengthTypeID == 0 and packetLength >= lengthOrNumOfSubPackets:
            done = True
    
    retVal = -1
    print(f'packetReturns {packetReturns}')
    if packetID == 0:
        retVal = sum(packetReturns)
    elif packetID == 1:
        retVal = multiplyList(packetReturns)
    elif packetID == 2:
        retVal = min(packetReturns)
    elif packetID == 3:
        retVal = max(packetReturns)
    elif packetID == 5:
        retVal = int(packetReturns[0] > packetReturns[1])
    elif packetID == 6:
        retVal = int(packetReturns[1] > packetReturns[0])
    elif packetID == 7:
        retVal = int(packetReturns[1] == packetReturns[0])
        
        
    return retVal, startbit


processPacket(b)
# vnumsum
```
