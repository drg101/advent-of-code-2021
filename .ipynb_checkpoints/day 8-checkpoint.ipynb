```python
i = open("input.txt", "r").read()
l = i.split('\n')
#l = [int(j) for j in l]
```

```python
l2 = [[k.split(' | ')[0],k.split(' | ')[1]] for k in l]
```

```python
u = [2,4,3,7]
m = {
    2:1,
    4:4,
    3:7,
    7:8
}

mapped = {
    
}

t = 0
digits=''
for o in l2:
    p = o[0]
    for y in p.split(' '):
        key = len(y)
        if key in m:
            digits+=f'{m[key]}' 
            mapped[m[key]] = y
    for y in p.split(' '):
        key = len(y)
        if key not in m:
            if key == 5:
                intcount = len(set(y).intersection(mapped[4]))
                #print(intcount)
                if intcount == 2:
                    mapped[2] = y
                    continue
                intcount = len(set(y).intersection(mapped[7]))
                if intcount == 2:
                    mapped[5] = y
                    continue
                intcount = len(set(y).intersection(mapped[1]))
                if intcount == 2:
                    mapped[3] = y
                    continue
#                 
            elif key == 6:
                #0,6,9
                intcount = len(set(y).intersection(mapped[1]))
                if intcount == 1:
                    mapped[6] = y
                    continue
    for y in p.split(' '):
        key = len(y)
        if key not in m and y not in mapped.values():
            if key == 6:
                intcount = len(set(y).intersection(mapped[3]))
                if intcount == 4:
                    mapped[0] = y
                else:
                    mapped[9] = y
    digString = ''
    rw = {value : key for (key, value) in mapped.items()}
    for digit in o[1].split(' '):
        for ke in rw:
            if len(set(digit).intersection(ke)) == len(digit) and len(digit) == len(ke):
                print(mapped)
                print(f'{digit} is a {rw[ke]}')
                digString += f'{rw[ke]}'
                break
            
    t+=int(digString)
print(mapped)
t
```
