```python
t = 0
pairMap = {
    '{': '}',
    '(': ')',
    '<': '>',
    '[': ']'
}

numIllegal = {
    '}': 0,
    ')': 0,
    '>': 0,
    ']': 0
}

cost = {
    ')': 1,
    ']': 2,
    '}': 3,
    '>': 4
}

scores = []
for line in l:
    stack = []
    myScore = 0
    corrupt = False
    for i in line:
        if i in pairMap:
            stack.append(i)
        if i in pairMap.values():
            z = stack.pop()
            if pairMap[z] != i:
                corrupt = True
                break
    if not corrupt:
        stackFix = ''
        while len(stack):
            repairWith = pairMap[stack.pop()]
            stackFix += repairWith
            myScore *= 5
            myScore += cost[repairWith]
        print("-")
        print(line)
        print(stackFix)
        scores.append(myScore)
    
            
sorted(scores)[len(scores)//2]
```
